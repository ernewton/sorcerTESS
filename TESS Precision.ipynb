```python
import math
import numpy as np
from NoiseModel import NoiseModel
import matplotlib.pyplot as plt
import palettable
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter

colors = palettable.colorbrewer.qualitative.Dark2_4.mpl_colors
import matplotlib 
matplotlib.rcParams.update({'font.size': 14})

```

```python
def optAper(Tmag):
    
    # this eq. is from the TIC paper. I think it returns to number of pixels in the aperture, which is what I want. 
    Cs = [-0.2592, 7.7410, -77.7918, 274.2898]
    p = np.poly1d(Cs)
    optAper = round(p(Tmag))
    if optAper < 4: 
        optAper = 4
        
    return optAper # noise calculation needs AREA in aperture (in pixels sq.)

print optAper(11)
```

```python
def calc_pp(Texp,             
            tmags = np.arange(10,16.5, 0.25), 
            contam = 0.,
):
    
    noise = np.zeros_like(tmags)
    
    for i, Tmag in enumerate(tmags):
        myNM = NoiseModel(Tmag, Texp*60, optAper(Tmag), fracflux=0.8, unit='percent')
        
        noise[i] = myNM.tot_noise()
        
    return noise
```

```python
plt.figure(figsize=(8,6))

tmags = np.arange(8,17.25,0.25)

x = calc_pp(2, tmags=tmags)
plt.plot(tmags, x, c=colors[0], label='2 min')

x = calc_pp(10, tmags=tmags)
plt.plot(tmags, x, c=colors[1], label='10 min')

x = calc_pp(30, tmags=tmags)
plt.plot(tmags, x, c=colors[2], label='30 min')

x = calc_pp(60, tmags=tmags)
plt.plot(tmags, x, c=colors[3], label='60 min')

plt.xlabel("TESS magnitude (mag)")
plt.ylabel("Error (%)")

plt.yscale('log')
plt.xticks(np.arange(8,18,1))
plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.1f'))

plt.grid(b=True, which='major', color='darkgray', linestyle='-')
plt.grid(b=True, which='minor', color='lightgray', linestyle='-')

plt.text(12.24, 0.015, 'Aperture size from Stassun et al. (2017)\nMinimum size = 4 sq. pix', fontsize=12)

plt.legend()
plt.title("SorcerTESS estimated precision in a single datum")
plt.tight_layout()

plt.savefig('outputs/SorcerTESS_pp_precision.png')
```
